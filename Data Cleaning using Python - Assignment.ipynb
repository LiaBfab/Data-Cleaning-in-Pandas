<h2>Question 6</h2>
<p> Unrealistic Values: Identify outliers in columns like </p>

```python
#The column with possible outlier is Age, especially where age falls below 10yrs. Let's check out the statistics
messyData.CusterAge.describe()
```

```python
q1=messyData.CusterAge.quantile(0.25)
q3=messyData.CusterAge.quantile(0.75)
IQR=q3-q1
IQR
outliers = messyData.CusterAge[(messyData.CusterAge<(q1-1.5*IQR)) | (messyData.CusterAge>(q3+1.5*IQR))].count()
outliers
```

<p><strong>Answer:</strong> There are no outliers in CustomerAge column</p>
